```python
import os
from pathlib import Path
from satisfying_sims.core import SimulationRecording
from satisfying_sims.themes.ice_cracks import IceCracksTheme
from satisfying_sims.core.recording import BodyStateSnapshot, BodyStaticSnapshot, ColliderSnapshot
from dataclasses import asdict
import matplotlib.pyplot as plt
```

```python
project_root = Path(os.getcwd()).parent
results_path = project_root / "results"
exp_name = "my_experiment"
pkl_path = results_path / exp_name / "recording.pkl.xz"
```

```python
recording = SimulationRecording.load(pkl_path)
```

```python
for id, body in recording.body_static.items():
    print(f'body static: {asdict(body)}')
```

```python
theme = IceCracksTheme()
theme.prepare_for_recording(body_ids=[0])
```

```python
count=0
body_state = BodyStateSnapshot(
    pos=(0.0, 0.0),
    vel=(0.0, 0.0),
    collision_count=count
)

body_static = BodyStaticSnapshot(
    id = 0, 
    mass = 1.0, 
    color = (0.0, 0.0, 1.0), 
    collider = ColliderSnapshot(
        kind = 'CircleCollider', 
        attrs = {'radius': 3.0})
)
```

```python
fig, ax = plt.subplots()
theme.draw_body(
        ax=ax,
        body_id=0,
        state=body_state,
        static=body_static)
```
