```python
from satisfying_sims.audio.event_rejection import RejectConfig, probability
import numpy as np
import matplotlib.pyplot as plt
```

```python
reject_cfg = RejectConfig(lam0=10, lam_max=100.0, k=2.0, p_min=0.0, p_max=1.0)
lam = np.logspace(-2, 4, 1000)
```

```python
p = probability(lam, cfg=reject_cfg)
```

```python
plt.plot(lam, p)
plt.xscale('log')
plt.ylim(-0.1, 1.1)
plt.xlabel('Event Rate')
plt.ylabel('Acceptance Probability')
```

```python
plt.plot(lam, p*lam)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Event Rate')
plt.ylabel('Accepted Event Rate')
plt.show()
```
