```python
from satisfying_sims.utils.preset_loader import load_preset
from pathlib import Path
import os
from pprint import pprint
```

```python
project_root = Path(os.getcwd()).parent
preset_path = project_root / "presets" / "preset.yaml"
```

```python
loaded = load_preset(preset_path)
```

```python
path = loaded.preset_path
resolved = loaded.resolved
files = loaded.loaded_files
```

```python
print("Preset Path:")
pprint(path, width=50)
print("\nResolved Paths:")
pprint(resolved, width=50)
print("\nLoaded Files:")
pprint(files, width=50)
```
