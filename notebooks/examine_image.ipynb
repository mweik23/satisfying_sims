```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import os
from pathlib import Path
```

```python
project_root = Path(os.getcwd()).parent
img_path = project_root / 'assets/sprites/big_aj/default.png'
```

```python
img = Image.open(img_path)
arr = np.array(img)
opaque_mask = arr[:, :, 3] > 25
plt.imshow(opaque_mask)
```

```python
max_x = np.max(np.where(opaque_mask)[1])
max_y = np.max(np.where(opaque_mask)[0])
min_x = np.min(np.where(opaque_mask)[1])
min_y = np.min(np.where(opaque_mask)[0])
print('max_x: ', max_x, 'max_y: ', max_y, 'min_x: ', min_x, 'min_y: ', min_y)
```
